```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 有些情况下需要指定encoding
因为数据集有一些不是Python默认utf-8编码的字符

```python
star_wars = pd.read_csv('D:/d/star_wars.csv', encoding='ISO-8859-1')
```

```python
star_wars.head(10)
```

### 应该留意数据集的RespondentID
毕竟每个应答者的ID是唯一，但是在上个cell的某些行中是空白的，所以要先删除RespondentID中无效的行。

```python
star_wars.columns
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
star_wars.head(10)
```

```python
cols = ['Have you seen any of the 6 films in the Star Wars franchise?',
       'Do you consider yourself to be a fan of the Star Wars film franchise?']
star_wars[cols[0]].value_counts()
```

```python
star_wars[cols[1]].value_counts()
```

### 关于cols包含的这两列代表的是Yes/No问题，也含有NaN（被访者选择不回答问题的地方）
我应该先通过value_counts（注意作用的对象是Series，不是DataFrame）来查看列中的所有唯一值，以及每个值出现的总次数。为了使数据更容易分析，可以通过Series.map（）函数将Yes/No处理成布尔，这样就不用这么麻烦去比较带字符串的rows。

```python
Yes_No = {
    'Yes': True,
    'No': False
}

for col in cols:
    star_wars[col] = star_wars[col].map(Yes_No)
star_wars
```

### 关于NaN值的调用书写格式
要调用的时候，书写格式为np.nan，而不能直接NaN，这样会报错

```python
to_boolean = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.nan: False
}
```

### 对于column的名字很长时，如下的方式比较麻烦，不宜采用，可以直接采用列名的切片方式如DataFrame.columns[3:9]，这样更简便易用
cols_6 = ['Which of the following Star Wars films have you seen? Please select all that apply.',
         'Unnamed: 4',
         'Unnamed: 5',
         'Unnamed: 6',
         'Unnamed: 7',
         'Unnamed: 8']

```python
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(to_boolean)
star_wars
```

### 关于DataFrame的列名
有些时候原始数据集的各个列名很长，不方便后续的数据处理，这时，可以先rename成好记，让人一目了然的名字，得注意rename的命令格式，和字典有点相似，不能混淆了

```python
star_wars = star_wars.rename(columns={'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
                                     'Unnamed: 4': 'seen_2',
                                     'Unnamed: 5': 'seen_3',
                                     'Unnamed: 6': 'seen_4',
                                     'Unnamed: 7': 'seen_5',
                                     'Unnamed: 8': 'seen_6'})
```

```python
star_wars
```

### 注意[9:15]的数字是什么类型
为了保险起见，先将它转化为数值类型，通过DataFrame.astype()转化

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'rank_1',
                                     'Unnamed: 10': 'rank_2',
                                     'Unnamed: 11': 'rank_3',
                                     'Unnamed: 12': 'rank_4',
                                     'Unnamed: 13': 'rank_5',
                                     'Unnamed: 14': 'rank_6'})
```

```python
star_wars
```

```python
rank_mean = star_wars[star_wars.columns[9:15]].mean()
rank_mean
```

```python
fig, ax = plt.subplots()
ranking = ax.bar(range(6), rank_mean, width=0.65, color='g',)
ax.set_ylabel('Rank Scores')
ax.set_title('The Mean of Each Rank')
ax.set_xticks(np.arange(6))
ax.set_xticklabels(['rank_1', 'rank_2', 'rank_3', 'rank_4', 'rank_5', 'rank_6'])
```

### 关于Rank的问题
到目前为止，已经清理好了数据，重新命名了几个columns，并计算了每部电影的平均排名。看起来Star Wars这个系列的早期几部影分要比新出的几部电影高得多。

```python
seen_sum = star_wars[star_wars.columns[3:9]].sum()
seen_sum
```

```python
fig, ax =plt.subplots()
ax.bar(np.arange(6), seen_sum, color='g', width=0.6)
ax.set_xticks(np.arange(6))
ax.set_xticklabels(['seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6'])
ax.set_title('The Sum of Each Seen')
ax.set_ylabel('Seen Sum Value')
ax.set_xlabel('Seen')
```

## 关于每部电影观看人数的问题
1means the film was the most favorite, and 6 means it was the least favorite
似乎原来的电影比新电影更受观众的观看。这加强了我们在排名中所看到的，早期的电影似乎更受欢迎。

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']
```

```python
males
```

```python
females
```

```python
M_rank_mean = males[males.columns[9:15]].mean()
M_rank_mean
```

```python
F_rank_mean = females[females.columns[9:15]].mean()
F_rank_mean
```

```python
M_seen_sum = males[males.columns[3:9]].sum()
M_seen_sum
```

```python
F_seen_sum = females[females.columns[3:9]].sum()
F_seen_sum
```

```python
w = 0.4
N = np.arange(6)
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
fig,ax = plt.subplots(figsize=(9,5))
M_rank = ax.bar(N, M_rank_mean, color='g', width=w)
F_rank = ax.bar(N+w, F_rank_mean, color='b', width=w)
ax.set_xticks(N+w/2)
ax.set_xticklabels(['rank_1','rank_2','rank_3','rank_4','rank_5','rank_6'])
ax.set_title("The Each Rank's Mean of Male and Female")
ax.set_ylabel('Rank Mean Value')
ax.legend((M_rank[0], F_rank[0]),('Male', 'Female'))   # 也可以这样，ax.legend(['M_rank', 'F_rank'])

#autolabel(M_rank)
#autolabel(F_rank)
```
